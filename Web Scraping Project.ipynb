SCRAPING MULTIPLE PAGES OF QUOTES TO SCRAPE WEBSITE

```python
for page in range(1,11):
  url="https://quotes.toscrape.com/page/"+str(page)+'/'
  req=requests.get(url)
  soup=BeautifulSoup(req.text,'html')

  trail_quote=soup.find_all('span',class_='text')
  quote_line=[title.text.strip() for title in trail_quote]
  trail_author=soup.find_all('small',class_='author')
  quote_author=[author.text.strip() for author in trail_author]
  list=[]
  for i in range(len(quote_author)):
       about='https://quotes.toscrape.com/author/'+quote_author[i].split(' ')[0]+'-'+quote_author[i].split(' ')[1]+'/'
       list.append(about)
  df_list=[]
  for i in range(1,11):
      df=pd.DataFrame(columns=['Inspirational_Quote','Author','About_Author'])
      df['Inspirational_Quote']=quote_line
      df['Author']=quote_author
      df['About_Author']=list
      df_list.append(df)
  Final_df=pd.concat(df_list,ignore_index=True)
```

SCRAPPING MULTIPLE PAGES IN "BOOKS TO SCRAPE" WEBSITE

```python
URL="https://books.toscrape.com/catalogue/page-1.html"
```

```python
data=[]
for pages in range(1,51):
  # Total 50 pages are in the website
  Link="https://books.toscrape.com/catalogue/page-"
  URL=Link+str(pages)+'.html'
  req=requests.get(URL)
  soup=BeautifulSoup(req.content,'html.parser')

  for sp in soup.find_all('li',class_="col-xs-6 col-sm-4 col-md-3 col-lg-3"):
    book_link="https://books.toscrape.com/catalogue/" + sp.find_all('a')[-1].get('href')
    title= sp.find_all('a')[-1].get('title')
    img_link="https://books.toscrape.com/" + sp.find('img').get('src')[3:]
    book_rating = (sp.find('p').get('class')[-1])
    price       = sp.find('p',class_='price_color').text[1:]
    stock       =  sp.find('p',class_ = 'instock availability').text.strip()
    data.append([title,book_rating,price,stock,book_link,img_link])
```

```python
df2=pd.DataFrame(data,columns=['title','book_rating','price','stock','book_link','img_link'])
```
